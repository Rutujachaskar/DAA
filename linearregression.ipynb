```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X=np.array([10,9,2,15,10,16,11,16]).reshape(-1,1)
Y=np.array([15,80,10,50,45,98,38,93])
```

```python
model=LinearRegression()
model.fit(X,Y)
```

```python
slope=model.coef_[0]
intercept=model.intercept_
```

```python
equation=f"Y={slope:.2f}X+{intercept:.2f}"
```

```python
print(f"Equation of best fit line is:{equation}")
```

```python
plt.scatter(X,Y,color='blue',label='Data Points')
plt.plot(X,model.predict(X),color='red',label='Best Fit Line')
plt.xlabel("No of hours spent driving")
plt.ylabel("Risk Score")
plt.title("Best fit line - Linear Regression")
plt.legend()
plt.show()
```
